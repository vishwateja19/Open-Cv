## Add two images

```python
lena_image=cv2.imread('./images/lena_color_512.tif',1)
monkey_image=cv2.imread('./images/mandril_color.tif',1)
print(lena_image.shape)
print(monkey_image.shape)

add_image=cv2.add(monkey_image,lena_image)
cv2.imshow('lena',lena_image)
cv2.imshow('monkey',monkey_image)
cv2.imshow('add',add_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### Giving weighted values to images while adding an image

```python
lena=cv2.imread('./images/lena_color_512.tif',1)
monkey=cv2.imread('./images/mandril_color.tif',1)

add_image=cv2.addWeighted(monkey,0.1,lena,0.9,1)
cv2.imshow('l',lena)
cv2.imshow('monk',monkey)
cv2.imshow('addweighted',add_image)
cv2.waitKey()
cv2.destroyAllWindows()
```
