- finding individual channels with opencv

```python
image_color = cv2.imread('./images/mandril_color.tif')  # loading with opencv
print(image_color.shape)

plt.subplot(2,2,1)
plt.title('B_Channel')
plt.imshow(image_color[:, : , 0],cmap=  'gray')

plt.subplot(2,2,2)
plt.title('G_Channel')
plt.imshow(image_color[:, : , 1],cmap=  'gray')

plt.subplot(2,2,3)
plt.title('R_Channel')
plt.imshow(image_color[:, : , 2],cmap=  'gray')

plt.subplot(2,2,4)
plt.title('original_image')
plt.imshow(image_color)

plt.show()
```
