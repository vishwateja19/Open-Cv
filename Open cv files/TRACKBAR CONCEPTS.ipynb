```python
## Importing packages

import numpy as np
import cv2
```

#### creating trackbar for some problem statements:

```python
def fun(x):
    print(x)
    
```

```python
cv2.namedWindow('track')
cv2.createTrackbar('basic','track',0,255,fun)

while True:
    black_image=np.zeros([512,512,3], dtype=np.float64)
    d=cv2.getTrackbarPos('basic','track')
    d=int(d)
    text=str(d)
    font=cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(black_image,text,(150,150),font,1,(255,255,255),1)
    cv2.imshow('check',black_image)
    if cv2.waitKey(10)& 0xFF==ord('q'):
        break
cv2.destroyAllWindows()        
    
```

```python
import cv2
import numpy as np
```

```python
def fun(x):
    print(x)
    
cv2.namedWindow('track')

cv2.createTrackbar('l_h','track',0,255,fun)
cv2.createTrackbar('l_s','track',0,255,fun)
cv2.createTrackbar('l_v','track',0,255,fun)


cv2.createTrackbar('h_h','track',255,255,fun)
cv2.createTrackbar('h_s','track',255,255,fun)
cv2.createTrackbar('h_v','track',255,255,fun)


while True:
    color_image=cv2.imread('./images/tulips.jpg',1)
    hsv_image=cv2.cvtColor(color_image,cv2.COLOR_BGR2HSV)
    
    l_h=int(cv2.getTrackbarPos('l_h','track'))
    l_s=int(cv2.getTrackbarPos('l_s','track'))
    l_v=int(cv2.getTrackbarPos('l_v','track'))
    
    h_h=int(cv2.getTrackbarPos('h_h','track'))
    h_s=int(cv2.getTrackbarPos('h_s','track'))
    h_v=int(cv2.getTrackbarPos('h_v','track'))
    
    lower_range=np.array([l_h,l_s,l_v])
    upper_range=np.array([h_h,h_s,h_v])
    
    mask=cv2.inRange(hsv_image,lower_range,upper_range)
    final=cv2.bitwise_and(color_image,color_image,mask=mask)
    
    cv2.imshow('track',final)
    cv2.imshow('mask',mask)
    if cv2.waitKey(10)& 0xFF==ord('q'):
        break
        
cv2.destroyAllWindows()        
```
